というわけで、納車後24時間を待たずして速攻内装を引っ剥がした。整備マニュアルは注文してあるのだがまだ入荷しておらず、後方にベットなどを作るにしても躯体の構造がわからないためである。

手始めに左側後方をベコベコ外す。内張りクリップで止まってるだけなので、プラヘラ使ってベコベコ外す。

外してみてびっくりしたのが、結構な間隔でタップ切ってある点。下記図の環状骨格部にいい感じの高さに多分M8でタップ切ってあるため、ここに2x4材なり止めてしまえば足作らなくてもベット作成できそうな感じ。（後述するが、強度的に多分中央に支柱立てる必要はアリ）この点は商業車ならでは。


![](https://toyota.jp/pages/contents/coaster/005_b_002/image/equip/safety/carlineup_coaster_equip_safety_3_01_pc.png)


[https://toyota.jp/pages/contents/coaster/005_b_002/image/equip/safety/carlineup_coaster_equip_safety_3_01_pc.png](https://toyota.jp/pages/contents/coaster/005_b_002/image/equip/safety/carlineup_coaster_equip_safety_3_01_pc.png)

あと外してみて気がついたのが床材。納車状態だと円凸付きのビニールマットが轢かれているが、なんで平らなんだろうと不思議に思ってたらその下にコンパネ埋まってた。。。更にその下をどうもケーブル通してるっぽい。この床材に関しては嫁氏より変えるべしとお達しが出ているため、ちょっとなんとかしようと思う。（全面にベッタリと粘着剤が塗布されてるので、早めに剥がさないと固着しそう）このコンパネ外さないと床面のネジ穴等が全くわからんので、ここは今週中に引っ剥がそうと思う。

というわけで、とりあえず今日はバラして具合を見た感じでした。

おまけで、嫁氏が納車時より欲しいと言っていた玄関マット的なものを、ニトリでベストサイズのものを見つけたため購入。我が家のバス、運転席以外土禁にしたのだが、流石に使い勝手考えるとスイングドア一段目は土足にしたほうが調子がいいため、その部分に敷くマットを探していたのだった。ジャストフィットだったのがコレ。
