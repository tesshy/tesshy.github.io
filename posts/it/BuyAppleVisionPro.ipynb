暫く書いてねーなと思い、Apple Vision Pro買ったのでその願末。

# 入手

2024-01-19、突如Apple Vision Proの発売がアナウンスされた。
私、2006年から林檎教の洗礼を受けており、最新のAppleデバイスは基本的に全部買う必要があった（ない）。

iPhone出るたびに買い、iPadが出た時はアメリカから個人輸入し、M1出た時もモリモリを。もう一体いくら使ったのか考えたくも無い（考えるつもりもない）。

さて困った、またアメリカだけ発売だ、俗にいうおま国だ。これは個人輸入すっかーと思ったら、X界隈見ると転送業者のアドレスは蹴られるケースが最近あるという情報。
まぁそん時はそん時だ！と思っていざ2024-02-09の予約日。まー繋がらない。カート入っても決済まで辿りつかない。

やっべーと思ってはや30分。納品予定日は3月に突入。あ、終わったわ・・・と思って諦めた。。。



んが、翌朝何気なくApple Storeアプリを見てみると在庫がある、ハワイに。
ここで何かが動いた、そうだ、ハワイに行こう！そう思い立ってPickUpのオーダーを入れた。

んが、翌々日何気なくApple Storeアプリを見てみると在庫がある、ラスベガスに。
ここで何かが動いた、そうだ、ラズベガスに行こう！そう思い立ってハワイのオーダーをキャンセルし、ラスベガスの店舗でPickUpオーダーを入れた。

おそらく多くの日本人は近場のハワイか、便数が多いロサンゼルス、サンプランシスコにピックアップに向かったであろう。
僕はおそらく日本人で唯一、ラスベガスに行ったものと思われる。（事実、日本人には現地でほぼ遭遇しなかった）

というわけで、ラスベガスの[Apple The Forum Shops](https://www.apple.com/retail/theforumshops/)にて無事にピックアップできたのであった。
ちなみにこの店舗、ラスベガスでは2番目の規模らしい。土地勘がある人向けにいうと、シーザーズパレスのショッピングモールの中にある。
ちょっと早く着いてしまったので時間潰しにスロットやって$100負けたのは秘密。

![店舗前で勝利の舞を披露](876AD79A-58EF-467D-9124-9A2AC02D2002_4_5005_c.jpeg)


詰まるところ何がしたいかというと、Appleが提唱する空間コンピューティングを体験したいだけである。マジでそれだけ。
それを行うには、ワイ、メガネが必要な人種である。そう、Optical Insertsが必要なのだ。
今回はこれに合わせて友人が働く眼科で綿密に視力測定をやってもらい、それを片手にアメリカの眼鏡処方箋をネットオーダーした。

が、これがまた厄介でZEISSから直送になる為店頭受け取りができないのだ。そして転送には対応していない。
そうなると、滞在中に受け取る必要がある。そう、ホテルに配送してそれを受け取るのだ。
過去Amazon.comの商品を受け取ろうと奮闘したことがあったのだが、ホテル側が超不親切で何もやってくれんかった。
（後から調べたが、フロントの人がわかってなかったらしい。仕組みとしてはある）
今回は事前にフロントに拙い英語で質問しておいたところ、ホテル内にあるFedexOfficeで一括してハンドリングしてるということがわかったので、
とりあえず滞在中に到着することを願った。そして、願いは叶った、帰国1日前に●それぐらいギリギリの戦いであった。
そう、私は勝利したのだ、完璧なApple Vision Proのセットを入手するということを！

![Apple Vision ProとOptical Inserts](AF81FBD8-8F70-468B-AB33-99F039BC9633_1_105_c.jpeg)

こうしてApple Vision Pro入手の目的は達せられた。
